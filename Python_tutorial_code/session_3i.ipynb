```python
# pandas
import pandas as pd # importing pandas with the alias pd

filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"
ci = pd.read_csv(filepath) # pd.read_excel() for reading excels
type(ci) # DataFrame pandas word for tabular data
# standard variable name for dataframes df
# child_data.info() # gives table info
# child_data.head(20) # top rows
# child_data.tail(10)  # bottom rows
#print(child_data)

ci.tail()
```

```python
ci["PersonBirthDate_dt"] = pd.to_datetime(ci["PersonBirthDate"], format="%Y-%m-%d")
ci["ExpectedPersonBirthDate_dt"] = pd.to_datetime(ci["ExpectedPersonBirthDate"], format="%Y-%m-%d")

ci["PersonDeathDate_dt"] = pd.to_datetime(ci["PersonDeathDate"], format="%Y-%m-%d", errors="coerce")
```

```python
# age is current date - birth date

collection_end = pd.to_datetime("31/03/2025", format="%d/%m/%Y")

ci["time_delta"] = collection_end - ci["PersonBirthDate_dt"]

ci['Age'] = ci["time_delta"] / pd.Timedelta(365.25, "d")

ci['Age'] = ci["Age"].astype(int) # converting to integers rounds down to closest int

ci['Age']
```
